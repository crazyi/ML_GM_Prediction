```python
import pandas as pd

# 生成两组示例数据
df = pd.read_csv('重力数据2整体测试结果.csv')

# 数据预处理
# scaler = StandardScaler() 
# df1[['longitude', 'latitude']] = scaler.fit_transform(df1[['longitude', 'latitude']])
# df2[['longitude', 'latitude']] = scaler.fit_transform(df2[['longitude', 'latitude']])
value = df['value'].values
RNNprediction = df['RNN3-prediction'].values
RFprediction = df['RF2-prediction'].values
SVMprediction = df['SVM-prediction'].values
krigingprediction = df['kriging2-prediction'].values

kriging1prediction = df['kriging1-prediction'].values
# 计算两组数据的差值
RNNdifferences = RNNprediction - value
RFdifferences = RFprediction - value
SVMdifferences = SVMprediction - value
krigingdifferences = krigingprediction - value
```

```python
import matplotlib.pyplot as plt

# 创建一个新的图形
fig = plt.figure(figsize=(10, 8))

# 创建子图1，位置为(0, 0)
plt.subplot(2, 2, 1)
plt.ylim(-60, 60)
plt.scatter(range(len(RNNdifferences)), RNNdifferences, color='k', s=10)
plt.xlabel('Index')
plt.ylabel('Difference')
plt.title('RNN')

# 创建子图2，位置为(0, 1)
plt.subplot(2, 2, 2)
plt.ylim(-60, 60)
plt.scatter(range(len(RFdifferences)), RFdifferences, color='k', s=10)
plt.xlabel('Index')
plt.ylabel('Difference')
plt.title('RF')

# 创建子图3，位置为(1, 0)
plt.subplot(2, 2, 3)
plt.ylim(-60, 60)
plt.scatter(range(len(SVMdifferences)), SVMdifferences, color='k', s=10)
plt.xlabel('Index')
plt.ylabel('Difference')
plt.title('SVM')

# 创建子图4，位置为(1, 1)
plt.subplot(2, 2, 4)
plt.ylim(-60, 60)
plt.scatter(range(len(krigingdifferences)), krigingdifferences, color='k', s=10)
plt.xlabel('Index')
plt.ylabel('Difference')
plt.title('Kriging')

# 调整子图之间的间距
plt.tight_layout()

# 保存图像
plt.savefig('散点图.png')

# 显示图形
plt.show()
```
