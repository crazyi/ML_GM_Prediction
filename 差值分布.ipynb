```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
# 生成两组示例数据
df = pd.read_csv('重力数据2整体测试结果.csv')
value = df['value'].values
RNNprediction = df['RNN3-prediction'].values
RFprediction = df['RF2-prediction'].values
SVMprediction = df['SVM-prediction'].values
krigingprediction = df['kriging2-prediction'].values

# 计算两组数据的差值
RNNdifferences = RNNprediction - value
RFdifferences = RFprediction - value
SVMdifferences = SVMprediction - value
krigingdifferences = krigingprediction - value
# 设置宋体字体路径
font_path = 'C:/Windows/Fonts/simsun.ttc'  # 请根据您的系统字体路径进行修改
font_prop_1 = FontProperties(fname=font_path, size=12)
font_prop_2= FontProperties(family='Times New Roman')
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
font_prop = FontProperties(family='Times New Roman')
plt.rcParams['font.size'] = 15
# # 示例数据：真实值和预测值
# y_true = np.array([100, 150, 200, 250, 300, 350, 400])
# y_pred = np.array([110, 140, 210, 240, 310, 330, 420])
# 
# # 计算残差（真实值 - 预测值）
# residuals = y_true - y_pred

# 创建样本索引（x 轴）
x = np.arange(len(RFdifferences))

# 绘制残差折线图
plt.figure(figsize=(8, 6))
plt.plot(x, RFdifferences, marker='o', linestyle='-', color='blue', label='Residuals')

# 添加中线 y=0，表示无残差的参考线
plt.axhline(0, color='red', linestyle='--', linewidth=2, label='Zero Residual')

# 图像标题和轴标签
plt.title('Residuals of RF')
plt.xlabel('Sample Index')
plt.ylabel('Residuals')

# 显示图例
plt.legend()

# 显示图像
plt.show()
```
